```python
import pandas as pd
import os
```

```python
data = pd.read_csv(filepath_or_buffer="Machine_data.csv")
```

```python
data.shape
```

```python
data = data.drop("score_difference", axis=1)
```

```python
X = data.drop("Winner", axis=1)
y = data["Winner"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
pd.set_option('display.max_columns', None)
list(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.15, random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
importances = rf.feature_importances_
importances
```

```python
sorted(zip(rf.feature_importances_, X.columns), reverse=True)
```

```python
game_prediction = pd.read_csv(filepath_or_buffer="prediction.csv")
game_prediction.head()
```
