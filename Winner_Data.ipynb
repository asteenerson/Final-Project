```python
# import data
import pandas as pd
import numpy as np
import glob
import os
import shutil
import datetime as dt
```

```python
path = r"C:\Users\Astee\Documents\School\Classes\23-Final-Project\NFL_ML\nflscrapR-data-master\games_data\regular_season" # use your path pbp-2019.csv

all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)

PA_add_df = frame
```

```python
Data = PA_add_df.filter(['type','game_id','home_team','away_team','week','season','home_score','away_score'])
```

```python
Home_team = PA_add_df.filter(['type','game_id','home_team','week','season','home_score'])
Away_team = PA_add_df.filter(['type','game_id','away_team','week','season','away_score'])
Score = PA_add_df.filter(['type','game_id','season','home_score','away_score'])
```

```python
Data['score_difference'] = Score['home_score'] - Score['away_score']
```

```python
Data['hometeam_result'] = np.where(Data['score_difference'] > 0, '1', '0')
Data['awayteam_result'] = np.where(Data['score_difference'] < 0, '1', '0')
```

```python
Data['Winner'] = np.where(Data['hometeam_result'] == '1', 'H', 'A')
```

```python
Home_team = Data.filter(['type','game_id','home_team','week','season','home_score','hometeam_result'])
Away_team = Data.filter(['type','game_id','away_team','week','season','away_score','hometeam_result'])
```

```python
home = Home_team.sort_values(by=['home_team','season'])
away = Away_team.sort_values(by=['away_team','season'])
```
